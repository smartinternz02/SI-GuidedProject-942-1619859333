```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)#
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'g')
plt.xlabel("linspace array")
plt.ylabel("powers of array")
plt.title("graph between x and y")
```

```python
plt.scatter(x,y,color="purple")
plt.xlabel("linspace array")
plt.ylabel("powers of array")
plt.title("graph between x and y")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g')
plt.subplot(2,2,4)
plt.plot(y,x,'r')
plt.subplot(2,2,2)
plt.scatter(x,y,color="orange")
plt.subplot(2,2,3)
plt.scatter(y,x,color="pink")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes.plot(x,y,'b')
axes2.scatter(x,y,color="green")
```

```python
plt.bar(x,y,color="yellow")
```

```python
plt.plot(x,y)
scale_factor=0.8
xmin,xmax=plt.xlim()
ymin,ymax=plt.ylim()
plt.xlim(xmin*scale_factor,xmax*scale_factor)
plt.ylim(ymin*scale_factor,ymax*scale_factor)
```

```python
x1=np.arange(0,10)
y1=np.arange(10,20)
```

```python
plt.plot(x,y,'r',label="xvsy")
plt.plot(y,x,'b',label="yvsx")
plt.plot(x1,y1,'g',label="x1vsy1")
plt.plot(y1,x1,'y',label="y1vsx1")
plt.legend()
```

```python
courses=["php","python","java","aiml","iot"]
number=[80,90,120,300,250]
plt.pie(number,labels=courses,autopct="%0.1f%%",colors=["orange","yellow","green","blue","purple"])
plt.title("pie chart")
plt.show()
```

```python
fig=plt.figure(figsize=(16,12))
plt.pie(number,labels=courses,autopct="%0.1f%%",colors=["orange","yellow","green","blue","purple"])
plt.title("pie chart")
plt.show()
```

```python
fig=plt.figure(figsize=(10,10))
plt.plot(x,y,"yellow")
plt.show()
```

```python
plt.bar(courses,number)
```

```python
fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(2,2,1)
ax2=fig.add_subplot(2,2,2)
ax3=fig.add_subplot(2,2,3)
ax4=fig.add_subplot(2,2,4)
ax1.plot(x,y)
ax2.scatter(y,x)
ax3.pie(number,labels=courses,autopct="%0.1f%%",colors=["orange","yellow","green","blue","purple"])
ax4.bar(courses,number,color="purple")
```
